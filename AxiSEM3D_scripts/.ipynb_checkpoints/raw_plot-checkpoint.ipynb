```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(dpi=200)

dim = 2
for idist, dist in enumerate([90,95,100,105,110,115,120,125,130,135]):
    seis_prem = np.loadtxt('output/stations/UV.D%d.RTZ.ascii' % dist)
    
    maxmax = np.max(np.abs(seis_prem[:, dim]))
    seis_prem[:, dim] /= maxmax 

    plt.plot(seis_prem[:,0],seis_prem[:,dim] + dist, c='b')

plt.xlim(0, 3600)
plt.show()
```
